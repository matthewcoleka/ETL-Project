```python
#Import dependencies
import requests
from config import api_key,app_id
from pprint import pprint
import pandas as pd
import numpy as np
import time as time





#Creating list of years and number of pages to pull from School Digger API
years = [2015,2016]
pages = np.arange(1, 11, 1)
data = []
```

```python
# Request school ranking data for years 2015-2016 from School digger API
for year in years:
    for page in pages:
        url = f'https://api.schooldigger.com/v1.1/rankings/schools/WA?year={year}&page={page}&perPage=50&appID={app_id}&appKey={api_key}'
        school_list = requests.get(url).json()
        data.append(dict(school_list))
        print(year)
        print(page)
        print(url)
        for i in range(len(school_list['schoolList'])):
            print(school_list['schoolList'][i]['schoolName']) 
        print(len(data))
        time.sleep(60)
        
```

```python
#Separating out 2015 and 2016 data, last pages of 2016 data incomplete due to API limitation
ranking_data_2015 = data[0:10]
ranking_data_2016 = data[10:20]
data[0]
```

```python
#Collcting list of schools and their id's and outting into a pandas dataframe
schools_list = []
for x in np.arange(0,10,1):
    for y in np.arange(0,50,1):
        #print(data[x]['schoolList'][y]['schoolName'], data[x]['schoolList'][y]['schoolid'] )
        schools_list.append({'school_id':(data[x]['schoolList'][y]['schoolid']),
                           'school_name': (data[x]['schoolList'][y]['schoolName']),
                           'school_district':(data[x]['schoolList'][y]['district']['districtName']),
                           'address':(data[x]['schoolList'][y]['address']['street']),
                           'city': (data[x]['schoolList'][y]['address']['city'])})
```

```python
#Transform school names/ids to pandas dataframe and push to mySQL database
schools_df = pd.DataFrame(schools_list, columns = ['school_id', 'school_name','school_district','address', 'city'])
schools_df.head()
```

```python
#Transform 2015 school rankings into list, and push to pandas dataframe
rankings2015_list = []


for x in np.arange(0,len(ranking_data_2015),1):
    for y in np.arange(0,50,1):
        rankings2015_list.append({'school_id':(ranking_data_2015[x]['schoolList'][y]['schoolid']),
                           'school_name': (ranking_data_2015[x]['schoolList'][y]['schoolName']),
                           'score_2015':(ranking_data_2015[x]['schoolList'][y]['rankHistory'][0]['averageStandardScore'])})
rankings2015_df = pd.DataFrame(rankings2015_list, columns = ['school_id', 'school_name', 'score_2015'])
rankings2015_df.head()
```

```python
#Transform 2015 school rankings into list, and push to pandas dataframe
rankings2016_list = []

for x in np.arange(0,len(ranking_data_2016),1):
    for y in np.arange(0,50,1):
        rankings2016_list.append({'school_id':(ranking_data_2016[x]['schoolList'][y]['schoolid']),
                           'school_name': (ranking_data_2016[x]['schoolList'][y]['schoolName']),
                           'score_2016':(ranking_data_2016[x]['schoolList'][y]['rankHistory'][0]['averageStandardScore'])})
rankings2016_df = pd.DataFrame(rankings2016_list, columns = ['school_id', 'school_name', 'score_2016'])
rankings2016_df.head()
```

```python
#Pulling in Vaccine Coverage data
vaccine_data_2015 = pd.read_csv('Data/CSV/WA_K_12_2015_2016.csv')
vaccine_data_2016 = pd.read_csv('Data/CSV/WA_K_12_2016_2017.csv')
vaccine_data_2015.head()
```

```python
#Create matching dataframes with lower values for identifier to correct for case sensitivity in matching
school_match = schools_df[['school_id','school_name','city','school_district']]
school_match['school_name_lower'] =school_match['school_name'].str.lower()
school_match['school_city_lower'] =school_match['city'].str.lower()
school_match['school_district_lower'] =school_match['school_district'].str.lower()
vaccine_data_2015['school_name_lower'] = vaccine_data_2015['School_Name'].str.lower()
vaccine_data_2015['school_city_lower'] = vaccine_data_2015['City'].str.lower()
vaccine_data_2016['school_name_lower'] = vaccine_data_2016['School_Name'].str.lower()
vaccine_data_2016['school_district_lower'] = vaccine_data_2016['School_District'].str.lower()
```

```python
#Merge 2015 Vax Data on school name and city to have school id's ,drop schools that are missing values
vax_school_2015 = pd.merge(school_match, vaccine_data_2015, how='left', left_on = ['school_name_lower','school_city_lower'], right_on=['school_name_lower','school_city_lower'], sort=True)
vax_school_2015
```

```python
#Merge 2015 Vax Data on school name and city to have school id's ,drop schools that are missing values
vax_school_2016 = pd.merge(school_match, vaccine_data_2016, how='left', left_on = ['school_name_lower','school_district_lower'], right_on=['school_name_lower','school_district_lower'], sort=True)
vax_school_2016
```

```python
#Clean datasets
vax_school_2015.drop(['school_name','city','school_district','school_name_lower','school_city_lower','school_district_lower'], axis=1,inplace=True)
vax_school_2016.drop(['school_name','city','school_district','school_name_lower','school_city_lower','school_district_lower'], axis=1,inplace=True)
```

```python
#Bring In SQL ALchemy Elements
from sqlalchemy import create_engine
import pymysql
pymysql.install_as_MySQLdb()
engine = create_engine("mysql://root:toor@localhost:3306/etl_db")
conn = engine.connect()
```

```python
#Push to mySQL
schools_df.to_sql('schools', con=engine, if_exists='replace')
rankings2015_df.to_sql('rankings_2015', con=engine, if_exists='replace')
rankings2016_df.to_sql('rankings_2016', con=engine, if_exists='replace')
vax_school_2015.to_sql('vax_schools_2015', con=engine, if_exists='replace')
vax_school_2016.to_sql('vax_schools_2016', con=engine, if_exists='replace')
```
